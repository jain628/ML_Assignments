```python
#Question:Click Here
#a.) Plot atleast 20 visualization for the weather dataset with different data points. 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather.head()
```

```python
#1.DISTPLOT
x = pd.Series(weather.MinTemp)
sns.distplot(x,color='green')
```

```python
#2
sns.distplot(x,hist=True,rug=True,kde=True,color='orange')
```

```python
#3
sns.distplot(x,hist=False,rug=False,kde=True)
```

```python
#4
sns.distplot(x,hist=False,rug=True,kde=True)
```

```python
#5
sns.distplot(x,hist=True,rug=True,kde=False)
```

```python
#6 KDEPLOT
sns.kdeplot(x,shade=True)
```

```python
#7
sns.kdeplot(x,color='red')
```

```python
#8
sns.kdeplot(x,bw = 0.2,label="bw=0.2")
```

```python
#9
sns.kdeplot(x,bw = 4,label="bw=4")
```

```python
#10
sns.kdeplot(x,bw = 4,label="bw=4")
sns.kdeplot(x,bw = 0.2,label="bw=0.2")
sns.kdeplot(x,label='original')
plt.legend()
```

```python
#11 PAIRPLOT
sns.pairplot(weather,vars=["MaxTemp","MinTemp","Rainfall","Evaporation"])
```

```python
#12
sns.pairplot(weather,vars=["MaxTemp","MinTemp","Rainfall","Evaporation"],hue='RainToday'\
            ,palette='Greens')
```

```python
#13
sns.pairplot(weather,vars=["MaxTemp","MinTemp","Rainfall","Evaporation"],diag_kind="kde",hue='RainToday'\
            ,palette='Greens')
```

```python
#14
sns.pairplot(weather,vars=["MaxTemp","MinTemp","Rainfall","Evaporation"],diag_kind='kde',hue='RainTomorrow'\
            ,palette='bone_r')
```

```python
#15 JOINTPLOT
sns.jointplot(data=weather,x='Evaporation',y='Rainfall',kind='scatter',color='r')
```

```python
#16
sns.jointplot(data=weather,x='Evaporation',y='Rainfall',kind='kde',color='r')
```

```python
#17
sns.jointplot(data=weather,x='Evaporation',y='Rainfall',kind='hex',color='r')
```

```python
#18 STRIPPLOT
sns.stripplot(x='RainToday',y='Humidity3pm',data = weather)
```

```python
#19
sns.stripplot(x='RainToday',y='Humidity3pm',data = weather,jitter=True)
```

```python
#20 SWARMPLOT
sns.swarmplot(x='WindGustDir',y='Humidity3pm',data = weather)
```

```python
#21 BOXPLOT
sns.boxplot(x='Cloud9am',y='Humidity3pm',data = weather)
```

```python
#22
sns.boxplot(x='Cloud3pm',y='Humidity3pm',data = weather,dodge=False,hue='RainToday')
```

```python
#23
sns.barplot(x='RainToday',y='Cloud9am',data = weather)
```

```python
#24 BARPLOT
sns.barplot(x='RainToday',y='Cloud9am',data = weather,hue='RainTomorrow')
```

```python
#25 COUNTPLOT
sns.countplot(x='RainToday',data = weather, palette='gist_heat')
```
