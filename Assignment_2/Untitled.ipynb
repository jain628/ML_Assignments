```python
import numpy as np
a = np.random.rand(1,10)
b = np.mean(a)
print("mean of array:",b)
```

```python
import numpy as np
a = np.random.rand(20,1)
b = np.std(a)
c = np.var(a)
print("standard-deviation:",b)
print("variance:",c)
```

```python
import numpy as np
a = np.random.rand(10,20)
b = np.random.rand(20,25)
c = np.matmul(a,b)
print("matrix multiplication:",c)
print("shape of matrix",np.shape(c))
print("sum of matrix",np.sum(c))
```
