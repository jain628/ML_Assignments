# 1) Import the required libraries (sklearn, pandas, numpy). Import datasets. Concatenate the datasets. Display first 5 rows.#

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1,df2],axis=0,sort=True)
df.head()
```

```python
df.shape
```

# 2) Remove NaN values from the column ‘Survived’. #

```python
df = df.dropna(subset=['Survived'])
df
```

# 3.Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not. 

```python
df['Has_cabin'] = df.apply(lambda row: 'yes' if pd.notnull(row.Cabin) else 'No', axis=1)
```

```python
df.head()
```

# 4) Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc. There should be five titles: Mr, Mrs, Master, Miss, and Special. #

```python
import re
df['Title'] = df.apply(lambda r:''.join(re.findall(', M[\w]+',r.Name)),axis=1)
```

```python
df.head()
```

```python
df
```

# 5. Impute the missing values using .fillna() method. #

```python
df = df.fillna(df.mean()).round(2)
```

```python
df.head()
```

```python
df
```

## Create two new columns “Categorical_age” and “Categorical_fare”, where you will put people that are within a certain range of age or fare in the same bin. ##

```python
df['Categorical_age'] = pd.qcut(df['Age'],q=4)
df['Categorical_fare'] = pd.qcut(df['Fare'],q=4)
```

```python
sns.countplot(data = df,x='Categorical_fare')
```

```python
sns.countplot(data = df,x='Categorical_age')
```

# 7) Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.#

```python
df3 = df.drop(columns = ['Cabin', 'Name', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'],axis=1)
df3
```

# 8) Transform all variables into numeric ones using get_dummies().#

```python
df = pd.get_dummies(df, columns=['Survived'])
```

```python
df
```

```python
df.columns
```

# 9. Now we need training set and testing set. Use train_test_split() to split your dataset. X(Features) will be the dataset values without the column “Survived” and y(Targets) will be your “Survived” column values.#

```python
x= np.array(df[['Age', 'Fare', 'Parch', 'PassengerId', 'Pclass',
     'SibSp']])
y= np.array(df['Survived_0.0'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=0)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

# 10) Use logistic regression by passing your X_train and y_train.#

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
y_pred = reg.predict(X_test)
```

# 11) Print the accuracy score.#

```python
score = reg.score(X_test, Y_test)
print(score)
```
